# **`Cleaning Data`**

## **Salary Data**

### **1. 'salary' Column**

After exploring the Salary column values, I found that the data is divided into 4 categories:

```python
oneValueSalaries = data.id[data['salary'].str.count('\£') == 1].count()
inRangeSalaries = data.id[data['salary'].str.count('\£') == 2].count()
notSpecifiedSalaries = data.id[data['salary'].str.contains('Not specified', case=False, na=False)].count()
nanSalaries = data.id[data['salary'].isna()].count()

print("'One value' salary counts",oneValueSalaries)
print("'Two values' salary counts",inRangeSalaries)
print("'Not specified' salary counts",notSpecifiedSalaries)
print("'Nan value' salary counts",nanSalaries)
print("Other values", data.id.count() - oneValueSalaries - inRangeSalaries - notSpecifiedSalaries - nanSalaries)
```

I'll divide my dataframe into 4 subsets for each salary category, and I'll work on every category independently. In the end I'll concatinate all the subsets again.

```python
# Divide the dataset into four subsets
one_value_subset = data[data['salary'].str.count('£') == 1]
in_range_subset = data[data['salary'].str.count('£') == 2]
not_specified_subset = data[data['salary'].str.contains('Not specified', case=False, na=False)]
nan_value_subset = data[data['salary'].isna()]

# Process 'one value' salaries
min_salaries_one_value = one_value_subset['salary'].str.extract(r'£([\d,]+(?:\.\d+)?)').replace(',', '', regex=True).astype(float)
max_salaries_one_value = min_salaries_one_value
one_value_subset['minSalary'] = min_salaries_one_value
one_value_subset['maxSalary'] = max_salaries_one_value

# Process 'in range' salaries
min_salaries_in_range = in_range_subset['salary'].str.extract(r'£([\d,]+(?:\.\d+)?)').replace(',', '', regex=True).astype(float)
max_salaries_in_range = in_range_subset['salary'].str.extract(r'to £([\d,]+(?:\.\d+)?)').replace(',', '', regex=True).astype(float)
in_range_subset['minSalary'] = min_salaries_in_range
in_range_subset['maxSalary'] = max_salaries_in_range

# Update 'Not specified' salaries
not_specified_subset['minSalary'] = None
not_specified_subset['maxSalary'] = None

# Update 'NaN' salaries
nan_value_subset['minSalary'] = None
nan_value_subset['maxSalary'] = None

# Concatenate the subsets
data = pd.concat([one_value_subset, in_range_subset, not_specified_subset, nan_value_subset], ignore_index=True)

# Display the relevant columns
data
```

Not like the `salary` column, the `addistionalSalaryInf` has a realy messy and a way different data formats, so getting clear data will be more challenging.

For this column, I will take the same approach, by dividing this dataframe into subsets and work on every one independently and concatinate all the subsets finally. But here I'm gonna divide the dataframe even more then before because here the values are varied a lot even in the same subset (for example, in 'In ranges' subset, some of the values use '-' and some of them use 'to').
So I need to explore the `additionalSalaryInf' data one more time to find some patterns that maybe can be used to build a Regax format.

### **4. Standardize the Salaries**

I decided to choose 'per hour' for all the salaries, so let's standardize all the salaries to 'per hour' wage.

```python
def convert_to_hourly_rate(row, salary_column):
    salary_type_mapping = {
        'per hour': 1,
        'per day': 1 / 8,  # Assuming 8 working hours per day
        'per week': 1 / 40,  # Assuming 40 working hours per week
        'per fortnight': 1 / 80,  # Assuming 80 working hours per fortnight
        'per month': 1 / (40 * 4),  # Assuming 40 working hours per week and 4 weeks per month
        'per year': 1 / (40 * 52)  # Assuming 40 working hours per week and 52 weeks per year
    }
    
    hourly_rate = row[salary_column]  # Default to the specified salary column
    if row['salaryType'] in salary_type_mapping:
        hourly_rate *= salary_type_mapping[row['salaryType']]
    
    if row['hours'] == 'Part Time':
        hourly_rate /= 2  # Assuming part-time jobs have half the working hours
    elif row['hours'] == 'Full Time' and row['salaryType'] == 'per year':
        hourly_rate /= 2080  # Assuming 40 hours per week and 52 weeks per year
    
    return hourly_rate

# Apply the function to overwrite 'minSalary' and 'maxSalary'
data['minSalary'] = data.apply(convert_to_hourly_rate, axis=1, salary_column='minSalary')
data['maxSalary'] = data.apply(convert_to_hourly_rate, axis=1, salary_column='maxSalary')
data
```

### **5. Finishing the min/max Salaries**

Now I'm going to check whether the min/max Salaryies are clean or not.

```python
data[(data['minSalary'] > data['maxSalary']) | ((data['minSalary'] == 0) & (data['maxSalary'] != 0)) | ((data['minSalary'] == 0) & (data['maxSalary'] == 0))]
```

```python
for index, row in data.iterrows():
    if not pd.api.types.is_number(row['minSalary']) or not pd.api.types.is_number(row['maxSalary']):
        # Delete the row if minSalary or maxSalary is not numeric
        data.drop(index, inplace=True)
    elif row['minSalary'] > row['maxSalary']:
        # Exchange minSalary and maxSalary if minSalary is greater than maxSalary
        data.at[index, 'minSalary'], data.at[index, 'maxSalary'] = row['maxSalary'], row['minSalary']
    elif row['minSalary'] == 0 and row['maxSalary'] != 0:
        data.at[index, 'minSalary'] = row['maxSalary']
    elif row['minSalary'] == 0 and row['maxSalary'] == 0:
        data.at[index, 'minSalary'], data.at[index, 'maxSalary'] = None, None

# Reset index after dropping rows
data.reset_index(drop=True, inplace=True)
```
